```python
import pandas as pd 
messages = pd.read_csv('SMSSpamCollection.tsv' , sep='\t' , names=['labels','message'])
```

```python
messages
```

```python
import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
```

### using porter Sttemmer 

```python
corpus=[]

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]' , ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

#### using lemmatization 

```python
cor2= []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]' , ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [wl.lemmatize(word,'v') for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    cor2.append(review)
```

```python
cor2
```

### Split the data into train and test sets 

```python
y = pd.get_dummies(messages['labels'])
y = y.iloc[:,0].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
X_train ,X_test , y_train,y_test = train_test_split(cor2 , y , test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
X_train.shape
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model = model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(y_test , y_pred))
```

## Created the TF-IDF model 

```python
y = pd.get_dummies(messages['labels'])
y = y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split 
X_train ,X_test , y_train,y_test = train_test_split(cor2 , y , test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
X_train =tf.fit_transform(X_train)
X_test = tf.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model =MultinomialNB()

model= model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test ,y_pred ))
```
