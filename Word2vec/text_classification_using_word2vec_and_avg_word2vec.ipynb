```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string
import nltk
import re
lemmtize = WordNetLemmatizer()
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
corpus = []
for i in range (0 , len(message)):
  review = re.sub('[^a-zA-Z]', ' ', message['message'][i])
  review = review.lower()
  review = review.split()
  review = [lemmtize.lemmatize(word) for word in review ]
  review = " ".join(review)
  corpus.append(review)
```

```python
[[i,j,k]for i , j , k in zip(list(map(len,corpus)),corpus , message['message']) if i < 1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
word = []
for sentence in corpus:
  sent_token = sent_tokenize(sentence)
  for sent in sent_token:
    word.append(simple_preprocess(sent))
```

```python
## lets train word2vec from scratch
model = gensim.models.Word2Vec(word)
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
  # remove out of vocabulary words and handle empty documents
  vecs = [model.wv[word] for word in doc if word in model.wv.index_to_key]
  if not vecs:
    return np.zeros(model.wv.vector_size) # Return a zero vector if the document is empty
  return np.mean(vecs, axis=0)
```

```python
y_pred= classifier.predict(X_test)
y_train_pred = classifier.predict(X_train)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_train, y_train_pred))
```
